```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load data files into Pandas DataFrames

#### Tetramers

# Generate GFN2 violin plots

#### Dimers

```python
data = [dimers_gfn['Neutral Runtime (s)'], dimers_gfn['Cation Runtime (s)'], dimers_gfn['Cation@Neutral Runtime (s)'], [i for i in dimers_gfn['Neutral@Cation Runtime (s)'] if i != 0]]

fig, (ax1) = plt.subplots(nrows=1, ncols=1, figsize=(7, 5), dpi = 200)
ax1.set_ylabel('CPU Time (s)', fontsize = 18)
parts = ax1.violinplot(data, points = 2000)

for pc in parts['bodies']:
    pc.set_facecolor('tab:orange')
    pc.set_edgecolor('orange')
    pc.set_alpha(0.5)

for partname in ('cbars','cmins','cmaxes'):
    vp = parts[partname]
    vp.set_edgecolor('darkorange')
    vp.set_linewidth(2)
    
ax1.xaxis.set_ticks_position('bottom')
ax1.set_xticks(np.arange(1, 5))
#plt.xticks(rotation=45)
ax1.set_xticklabels(['Neutral','Cation','Cation@Neutral','Neutral@Cation'], fontsize = 14, rotation=45, ha = 'right')
plt.yscale('log')
plt.tight_layout()
#plt.savefig('Dimers_GFN2_runtime_violin.pdf')
plt.show()
```

# Generalt comparison graph

#### GFN2

```python
fig=plt.figure(figsize=(7, 5), dpi= 200, facecolor='w', edgecolor='k')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
x = ['Dimers','Tetramers','Hexamers']
plt.errorbar(x,[dimers_gfn['Cation Runtime (s)'].mean(),tetramers_gfn['Cation Runtime (s)'].mean(),hexamers_gfn['Cation Runtime (s)'].mean()],
            yerr=[dimers_gfn['Cation Runtime (s)'].sem(),tetramers_gfn['Cation Runtime (s)'].sem(),hexamers_gfn['Cation Runtime (s)'].sem()],
            capsize = 5,
            marker = 'o',
            linestyle = '--',
            c = 'tab:red',
            label='Cation')
plt.errorbar(x,[dimers_gfn['Neutral Runtime (s)'].mean(),tetramers_gfn['Neutral Runtime (s)'].mean(),hexamers_gfn['Neutral Runtime (s)'].mean()],
            yerr=[dimers_gfn['Neutral Runtime (s)'].sem(),tetramers_gfn['Neutral Runtime (s)'].sem(),hexamers_gfn['Neutral Runtime (s)'].sem()],
            capsize = 5,
            marker = 'o',
            linestyle = '--',
            c = 'tab:blue',
            label='Neutral')
plt.errorbar(x,[dimers_gfn['Neutral@Cation Runtime (s)'].mean(),tetramers_gfn['Neutral@Cation Runtime (s)'].mean(),hexamers_gfn['Neutral@Cation Runtime (s)'].mean()],
            yerr=[dimers_gfn['Neutral@Cation Runtime (s)'].sem(),tetramers_gfn['Neutral@Cation Runtime (s)'].sem(),hexamers_gfn['Neutral@Cation Runtime (s)'].sem()],
            capsize = 5,
            marker = 'o',
            linestyle = '--',
            c = 'tab:orange',
            label='Neutral@Cation')
plt.errorbar(x,[dimers_gfn['Cation@Neutral Runtime (s)'].mean(),tetramers_gfn['Cation@Neutral Runtime (s)'].mean(),hexamers_gfn['Cation@Neutral Runtime (s)'].mean()],
            yerr=[dimers_gfn['Cation@Neutral Runtime (s)'].sem(),tetramers_gfn['Cation@Neutral Runtime (s)'].sem(),hexamers_gfn['Cation@Neutral Runtime (s)'].sem()],
            capsize = 5,
            marker = 'o',
            linestyle = '--',
            c = 'tab:purple',
            label='Cation@Neutral')
plt.grid(axis='y', alpha = 0.2)
plt.legend()
#plt.title('CPU Runtime', fontsize = 18)
plt.ylabel('Mean CPU Runtime (s)', fontsize = 18)
plt.tight_layout()
#plt.savefig('Mean_GFN2_Runtime.png')
#plt.savefig('Mean_GFN2_Runtime.pdf')
plt.show()
```
