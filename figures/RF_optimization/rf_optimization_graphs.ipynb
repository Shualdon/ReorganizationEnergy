```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
rf_opt_df = pd.read_csv('RF_optimization_results4.csv')
```

```python
rf_opt_df['Score/Runtime'] = rf_opt_df['Score'] / rf_opt_df['Runtime']
```

```python
rf_opt_df
```

```python
fig=plt.figure(figsize=(6, 5), dpi= 200, facecolor='w', edgecolor='k')
plt.grid(color='k', ls = '-.', lw = 0.1)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.plot(rf_opt_df['Trees'],rf_opt_df['Score'], linewidth = 3, color = 'tab:blue')
plt.xlabel('Number of Trees', fontsize = 18)
plt.ylabel('Score', fontsize = 18)
plt.tight_layout()
plt.savefig('rf_optimization_score_vs_trees.pdf')
plt.savefig('rf_optimization_score_vs_trees.png')
plt.show()
```

```python
fig=plt.figure(figsize=(6, 5), dpi= 200, facecolor='w', edgecolor='k')
plt.grid(color='k', ls = '-.', lw = 0.1)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.plot(rf_opt_df['Trees'],rf_opt_df['Runtime'], linewidth = 3, color = 'tab:red')
plt.xlabel('Number of Trees', fontsize = 18)
plt.ylabel('Runtime (s)', fontsize = 18)
plt.tight_layout()
plt.savefig('rf_optimization_runtime_vs_trees.pdf')
plt.savefig('rf_optimization_runtime_vs_trees.png')
plt.show()
```

```python
fig=plt.figure(figsize=(6, 5), dpi= 200, facecolor='w', edgecolor='k')
plt.grid(color='k', ls = '-.', lw = 0.1)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.plot(rf_opt_df['Trees'],rf_opt_df['Score/Runtime'], linewidth = 3, color = 'tab:orange')
plt.xlabel('Number of Trees', fontsize = 18)
plt.ylabel('Score / Runtime ($s^{-1}$)', fontsize = 18)
plt.tight_layout()
plt.savefig('rf_optimization_score-runtime_vs_trees.pdf')
plt.savefig('rf_optimization_score-runtime_vs_trees.png')
plt.show()
```

```python
fig=plt.figure(figsize=(6, 5), dpi= 200, facecolor='w', edgecolor='k')
plt.grid(color='k', ls = '-.', lw = 0.1)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.plot(rf_opt_df['Trees'],rf_opt_df['TotalRMSE'], linewidth = 3, color = 'tab:cyan')
plt.xlabel('Number of Trees', fontsize = 18)
plt.ylabel('RMSE (eV)', fontsize = 18)
plt.tight_layout()
plt.savefig('rf_optimization_rmse_vs_trees.pdf')
plt.savefig('rf_optimization_rmse_vs_trees.png')
plt.show()
```
