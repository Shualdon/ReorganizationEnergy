```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error
from math import sqrt
import numpy as np
```

```python
all_df = pd.read_csv('../../data/DFT_GFN2_angle_bond_correlations.csv')
```

```python
fig=plt.figure(figsize=(6,6), dpi= 200, facecolor='w', edgecolor='k')
plt.grid(color='k', ls = '-.', lw = 0.1)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlim(90,182)
plt.ylim(90,182)
reg_x = np.arange(90,181,0.1)
sct = plt.scatter(all_df['GFN2 Neutral Angle'], all_df['DFT Neutral Angle'], color = 'tab:orange', edgecolors = 'none', alpha = 0.3)
slope, intercept, r_value, p_value, std_err = linregress(all_df['GFN2 Neutral Angle'], all_df['DFT Neutral Angle'])
plt.plot(reg_x,slope*reg_x + intercept, c = 'tab:blue', lw = 3, alpha = 0.7, ls = '--')
plt.ylabel('B3LYP Neutral Dihedral Angle ()', fontsize=18)
plt.xlabel('GFN2 Neutral Dihedral Angle ()', fontsize=18)
plt.legend([sct],['$R^2$ = {:.2f}'.format(r_value**2)], loc = 4)
#plt.title('Neutral Angle')
plt.tight_layout()
#plt.savefig('neutral_angle_correlation.pdf')
#plt.savefig('neutral_angle_correlation.png')
plt.show()
```
