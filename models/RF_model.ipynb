```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import matplotlib.patches
from matplotlib import cm
from IPython.display import Image, display
import numpy as np
import sys
from scipy.stats import linregress
from sklearn.metrics import mean_squared_error
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from math import sqrt
import math
from functools import reduce
```

# Load Data

```python
data = pd.read_csv('../data/reorganization_energy_data_ECFP4_clean.csv')
data_df = data[data['DFT λ'] > 0].reset_index(drop = False)
data_df = data_df[data_df['DFT λ'] < 2].reset_index(drop = False)
data_df.drop(['level_0','index'], axis = 1, inplace = True)
```

```python
print('Number of features (m):',len(data_df.columns))
print('Number of data points (n):',len(data_df))
```

```python
data_df.describe()
```

### Feature List

```python
print(list(dict.fromkeys([x.split('_')[0] for x in data_df.drop(['DFT λ'], axis = 1).columns])))
```

# Create Training and Test Sets

```python
X = data_df.drop(['DFT λ'], axis = 1)
y = data_df['DFT λ']

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.15, random_state=420)
n_cols = len(X.columns)
```

# Train Random Forrest Regressoion Model

```python
regr = RandomForestRegressor(n_estimators = 50, max_features = None, min_samples_split=2, random_state=0, verbose = 0)
```

```python
%%time
regr.fit(train_X, train_y)
```

```python
regr.score(test_X, test_y)
```

## Most Important Features

```python
for x,y in sorted(zip(data_df.columns, regr.feature_importances_), reverse = True, key = lambda x: x[1])[:10]: print('{}: {:.2f}%'.format(x,y*100))
```

```python
features = X.columns
importances = regr.feature_importances_
indices = np.argsort(importances)
num_features = 10 

my_cmap = cm.get_cmap('jet')

plt.figure(figsize=(10,6), dpi = 200)
#plt.title('Feature Importances')

# only plot the customized number of features
plt.barh(range(num_features), importances[indices[-num_features:]], color = my_cmap(np.arange(0,1,0.1)), align='center')
plt.yticks(range(num_features), [features[i] for i in indices[-num_features:]], fontsize = 16)
plt.xlabel('Relative Importance', fontsize = 20)
plt.xticks(fontsize = 16)
plt.tight_layout()
#plt.savefig('rf_feature_importance.pdf')
plt.show()
```

# Prediction

```python
test_y_predictions = np.asarray(regr.predict(test_X))
```

```python
pdf = test_X[['Oligomer Length','Monomer 1','Monomer 2']].copy(deep=True)
pdf['DFT λ'] = np.asarray(test_y)
pdf['Trained λ'] =  [x for x in list(test_y_predictions)]
pdf['PctDiff'] = (abs(pdf['DFT λ']-pdf['Trained λ'])/pdf['DFT λ'])*100
```

```python
print('In the test set there are...')
print('Tetramers: ' + str(len(pdf[pdf['Oligomer Length'] == 4])))
print('Hexamers: ' + str(len(pdf[pdf['Oligomer Length'] == 6])))
```

```python
fig=plt.figure(figsize=(8, 6), dpi= 300, facecolor='w', edgecolor='k')
plt.grid(color='k', ls = '-.', lw = 0.1)
plt.xlim(0,1.2)
#plt.ylim(0,1)
reg_x = np.arange(0,1.3,0.1)

if len(pdf[pdf['Oligomer Length'] == 4]) > 0:
    tetra_sct = plt.scatter(pdf[pdf['Oligomer Length'] == 4]['Trained λ'], pdf[pdf['Oligomer Length'] == 4]['DFT λ'], s = 30, c = 'tab:blue', edgecolors = 'none', alpha = 0.6, label = "Tetramers")
    slope, intercept, t_r_value, p_value, std_err = linregress(pdf[pdf['Oligomer Length'] == 4]['Trained λ'].astype(np.float),pdf[pdf['Oligomer Length'] == 4]['DFT λ'])
    plt.plot(reg_x,slope*reg_x + intercept, c = 'royalblue', lw = 1, alpha = 0.7)
    t_rmse = sqrt(mean_squared_error(pdf[pdf['Oligomer Length'] == 4]['Trained λ'],pdf[pdf['Oligomer Length'] == 4]['DFT λ']))
    
if len(pdf[pdf['Oligomer Length'] == 6]) > 0:
    hex_sct = plt.scatter(pdf[pdf['Oligomer Length'] == 6]['Trained λ'], pdf[pdf['Oligomer Length'] == 6]['DFT λ'], s = 30, c = 'tab:red', edgecolors = 'none', alpha = 0.6,  label = "Hexamers")
    slope, intercept, h_r_value, p_value, std_err = linregress(pdf[pdf['Oligomer Length'] == 6]['Trained λ'].astype(np.float),pdf[pdf['Oligomer Length'] == 6]['DFT λ'])
    plt.plot(reg_x,slope*reg_x + intercept, c = 'tab:red', lw = 1, alpha = 0.8)
    h_rmse = sqrt(mean_squared_error(pdf[pdf['Oligomer Length'] == 6]['Trained λ'],pdf[pdf['Oligomer Length'] == 6]['DFT λ']))

slope, intercept, total_r_value, p_value, std_err = linregress(pdf['Trained λ'].astype(np.float),pdf['DFT λ'])
total_plt, = plt.plot(reg_x,slope*reg_x + intercept, c = 'gray',ls = '--', lw = 2, alpha = 0.8, label = 'Total')

total_rmse = sqrt(mean_squared_error(pdf['Trained λ'],pdf['DFT λ']))
total = matplotlib.patches.Rectangle((0,0), 1, 1, fill=False, edgecolor='none', visible=False)
plt.xlabel('Trained λ (eV)', fontsize = 18)
plt.ylabel('B3LYP λ (eV)', fontsize = 18)
plt.xticks(fontsize = 16)
plt.yticks(fontsize = 16)

plt.legend([tetra_sct, hex_sct, total], 
           ['Tetramers\n$R^2$ = {:.3f}\nRMSE = {:.3f} eV'.format(t_r_value ** 2, t_rmse), 
            'Hexamers\n$R^2$ = {:.3f}\nRMSE = {:.3f} eV'.format(h_r_value ** 2, h_rmse),
            'Total\n$R^2$ = {:.3f}\nRMSE = {:.3f} eV'.format(total_r_value ** 2, total_rmse),], 
           loc = 8, ncol=3, bbox_to_anchor=(0.49, -0.33), markerscale = 1.5, fontsize = 12, frameon = True)
plt.tight_layout()
#plt.savefig('RF_test_predictions.pdf')
plt.show()
```
